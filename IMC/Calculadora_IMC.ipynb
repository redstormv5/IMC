```python
"""
Calculadora de IMC

Este programa recebe peso e altura do usuário,
calcula o IMC e retorna a classificação de acordo
com o gênero informado.

Autor: Lucas Noronha
Data: Junho/2025
"""
```

```python
# Calcula o IMC (Índice de Massa Corporal) com base no peso e na altura
def calculo(peso, altura):
    imc = round((peso / altura ** 2),1)
    return imc

# Define e exibe a categoria do IMC com base no gênero informado
def mulher(imc):
    if imc < 19:
        print(f"IMC {imc}: Abaixo do Peso")
    elif imc >=19 and imc < 24:
        print(f"IMC {imc}: Normal")
    elif imc >= 24 and imc < 29:
        print(f"IMC {imc}: Abesidade Leve")
    elif imc >= 29 and imc < 39:
        print(f"IMC {imc}: Obesidade Moderada")
    else:
        print(f"IMC {imc}: Obesidade Mórbida")

def homem(imc):
    if imc < 20:
        print(f"IMC {imc}: Abaixo do Peso")
    elif imc >=20 and imc < 25:
        print(f"IMC {imc}: Normal")
    elif imc >= 25 and imc < 30:
        print(f"IMC {imc}: Abesidade Leve")
    elif imc >= 30 and imc < 40:
        print(f"IMC {imc}: Obesidade Moderada")
    else:
        print(f"IMC {imc}: Obesidade Mórbida")

# Exibe o resultado de acordo com o gênero
def resultado(imc,genero):
    if genero == 1:
        homem(imc)
    else:
        mulher(imc)
```

```python
# Inicia o loop para permitir múltiplos cálculos de IMC
while True:

    try:

        # Solicita os dados do usuário: gênero, altura e peso
        genero = int(input("Insira 1 ou 2\nQual seu gênero? 1 - Masculino ou 2 - Feminino:"))
        if genero != 1 and genero != 2:
            raise ValueError
        altura = float(input("Digite a altura em metros. Ex: 1.80\nQual a sua altura?"))
        peso = float(input("Digite o peso em kg. Ex: 70.5\nQual o seu peso?"))

    #Caso o usúario digite algo errado, voltara ao inicio do programa
    except:
      print("Informe dados válidos")
      continue

    # Executa o cálculo do IMC e exibe o resultado com base no gênero
    imc = calculo(peso,altura)
    resultado(imc,genero)

    # Pergunta ao usuário se deseja realizar um novo cálculo
    final = int(input("Deseja realizar outro cálculo?\n1 - Sim ou 2 - Não"))

    # Encerra o loop se a resposta for 2 (não)
    if final == 2:
        break
```
